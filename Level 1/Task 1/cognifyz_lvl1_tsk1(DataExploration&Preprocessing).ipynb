# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the Dataset

```python
# Load the Dataset
df = pd.read_csv("Dataset .csv")
```

```python
# Display the first few rows to check the data
df.head()
```

# Explore the Dataset

```python
# Check the shape of the dataset
df.shape
```

# Check for Missing Values

```python
# Check for missing values in each columns
df.isnull().sum()
```

```python
df[df["Cuisines"].isnull()]
```

```python
df['Cuisines'] = df.groupby("City")["Cuisines"].apply(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown"))
```

# Check Data Types

```python
# Check the data type of each clumn
df.dtypes
```

```python
# Convert categorical Yes/No columns to Boolean (True/False)
bool_cols = ["Has Table booking", "Has Online delivery", "Is delivering now", "Switch to order menu"]
df[bool_cols] = df[bool_cols].applymap(lambda x: True if x== "Yes" else False)
```

```python
# Convert categorical numerical columns to 'category' data type
df["Price range"] = df["Price range"].astype("category")
df["Rating color"] = df["Rating color"].astype("category")
```

```python
# Convert 'Country Code' from integer to string
df["Country Code"] = df["Country Code"].astype(str)
```

# Analyze the Target Variable ("Aggregate rating")

```python
# Plot histogram for the target variable 'Aggregate rating'
df['Aggregate rating'].hist()
plt.title('Aggregate Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

# Check for Class Imbalance

```python
# Check for class imbalance
print(df['Aggregate rating'].value_counts())
```
